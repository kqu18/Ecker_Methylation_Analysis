```python
from wmb import brain, cemba, mm10
import wmb
from ALLCools.mcds import MCDS
import pandas as pd
import pathlib
import scanpy as sc
from qnorm import quantile_normalize
import numpy as np
import matplotlib.pyplot as plt
```

```python
meta = pd.read_csv('0523_seurat_metadata.csv', index_col = 0)

var_dim = 'gene'
mc_type = 'CHN'
```

```python
mcds = MCDS.open(mcds_paths = 'gene_mcds', 
                 use_obs = meta.index, 
                 var_dim = var_dim).sel(mc_type = f'{mc_type}')
```

```python
mcds.add_mc_frac(
    var_dim = var_dim, 
    normalize_per_cell=True,  # after calculating mC frac, per cell normalize the matrix
    clip_norm_value=10  # clip outlier values above 10 to 10
)
```

```python
frac_df =mcds['gene_da_frac'].to_pandas()
```

```python
frac_df['cluster'] = meta['seurat_clusters']
cluster_frac_df = frac_df.groupby('cluster').sum().T
cluster_frac_df.to_csv(f'{mc_type}_cluster_mc_frac_matrix.csv')
```
